```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (10,8)
```

```python
plt.rcParams["figure.figsize"] = (10,8)
datasim1 = np.loadtxt("sim1.txt")
[t, lp, lp_a, lp_b, ld, ld_a, ld_b] = datasim1.T
plt.plot(t, ld_a, color = 'r', linestyle = '--', label = 'min L_DA(1)')
plt.plot(t, ld_b, color = 'c', linestyle = '-.', label = 'min L_DA(2)')
plt.plot(t, lp, color = 'g', label = 'min L_P')
plt.axvspan(600, 1200, alpha=0.2, color='grey')
#plt.axvspa
#TODO: add partition 2
plt.xlabel('Time[s]')
plt.ylabel('Ledger Length[blks]')
plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10,8)
data = np.loadtxt("sim3.txt")
[t, lp, lda, l_awake, l_asleep] = data.T
#Sim3
plt.plot(t, lda, color = 'r', label = 'min L_DA')
plt.plot(t, lp, color = 'b', label = 'min L_P')
plt.xlabel('Time[s]')
plt.ylabel('Ledger Length[blks]')
plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (10,3)
tshld = 67
min_awake = min(l_awake)
max_awake = max(l_awake)

plt.plot(t, l_awake, label = "Honest Awake", color = 'black')
plt.plot(t, [tshld] * len(t), label = "threshold", color = 'green')
plt.plot(t, [min_awake] * len(t), label = "min", color = 'red')
plt.plot(t, [max_awake] * len(t), label = "max", color = 'red')
plt.xlabel('Time[s]')
plt.ylabel('Awake honest')
plt.legend()
plt.show()
```
