```python
plt.rcParams["figure.figsize"] = (10,3)
tshld = 67
min_awake = min(l_awake)
max_awake = max(l_awake)

arrs = []
curr = (-1, -1)

for i, l in enumerate(l_awake):
    if l >= 67:
        if curr == (-1, -1):
            curr = (i, i)
        else:
            curr = (curr[0], i)
    else:
        arrs.append(curr)
        curr = (-1, -1)
        
if curr != (-1, -1):
    arrs.append((curr[0], len(arrs)))
    
plt.plot(t, l_awake, label = "Honest Awake", color = 'black')
plt.plot(t, [tshld] * len(t), label = "threshold", color = 'green')
plt.plot(t, [min_awake] * len(t), label = "min", color = 'red')
plt.plot(t, [max_awake] * len(t), label = "max", color = 'red')
plt.xlabel('Time[s]')
plt.ylabel('Awake honest')
plt.legend(loc=0, prop={'size': 15})

print(arrs)

for st, end in arrs:
    if st != -1:
        plt.axvspan(15*st, 15*end, alpha=0.1, color='green')

plt.show()
```
